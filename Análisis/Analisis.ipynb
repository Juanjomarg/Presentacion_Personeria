### Presupuesto

#### Presupuesto por sector

```python
respuestas_raw=pd.read_excel(f'{ruta}/raiz/respuestas_2023.xlsx').sort_values(by='sector')
respuestas_raw['funcionamiento']=respuestas_raw[['p3_val_1', 'p3_val_2']].mean(axis=1)
respuestas_raw['funcionamiento_innovacion']=respuestas_raw[['p4_val_1', 'p4_val_2']].mean(axis=1)

respuestas_raw['inversion']=respuestas_raw[['p5_val_1', 'p5_val_2']].mean(axis=1)
respuestas_raw['inversion_innovacion']=respuestas_raw[['p6_val_1', 'p6_val_2']].mean(axis=1)

respuestas_raw['general']=respuestas_raw[['funcionamiento', 'inversion']].mean(axis=1)
respuestas_raw['general_innovacion']=respuestas_raw[['funcionamiento_innovacion', 'inversion_innovacion']].mean(axis=1)

respuestas_raw.to_excel(f'{ruta}/raiz/respuestas_2023.xlsx',index=False)

presupuesto_funcionamiento_sector=respuestas_raw.groupby('sector')['funcionamiento'].mean()
presupuesto_funcionamiento_innovacion_sector=respuestas_raw.groupby('sector')['funcionamiento_innovacion'].mean()

presupuesto_inversion_sector=respuestas_raw.groupby('sector')['inversion'].mean()
presupuesto_inversion_innovacion_sector=respuestas_raw.groupby('sector')['inversion_innovacion'].mean()

presupuesto_general_sector=respuestas_raw.groupby('sector')['general'].median()
presupuesto_general_innovacion_sector=respuestas_raw.groupby('sector')['general_innovacion'].median()

respuestas_raw.sample(2)
```

```python
entidades=respuestas_raw[['sector','entidad','general']]

entidades_filtered = entidades[entidades['general'] > 0]
entidades_filtered

fig = px.treemap(entidades_filtered, path=['sector','entidad'], values='general',width=800, height=450)
fig.update_traces(marker=dict(cornerradius=5))
fig.show()

fig.write_html(f'{ruta}/graphs/entidades.html')
```

```python
funcionamiento = px.bar(
    respuestas_raw,
    x='sector',
    y='funcionamiento',
    color='sector',  # Use 'entidad' as the color variable
    labels={'sector': 'Sector', 'funcionamiento': 'Funcionamiento', 'entidad': 'Entidad'},
    title="Presupuesto de funcionamiento por Sector",
    hover_data=['entidad']  # Show 'entidad' on hover
)

funcionamiento_innovacion = px.bar(
    respuestas_raw,
    x='sector',
    y='funcionamiento_innovacion',
    color='sector',  # Use 'entidad' as the color variable
    labels={'sector': 'Sector', 'funcionamiento_innovacion': 'Funcionamiento Innovación', 'entidad': 'Entidad'},
    title="Presupuesto de funcionamiento para innovación por Sector",
    hover_data=['entidad']  # Show 'entidad' on hover
)

inversion = px.bar(
    respuestas_raw,
    x='sector',
    y='inversion',
    color='sector',  # Use 'entidad' as the color variable
    labels={'sector': 'Sector', 'inversion': 'Inversion', 'entidad': 'Entidad'},
    title="Presupuesto de inversión por Sector",
    hover_data=['entidad']  # Show 'entidad' on hover
)

inversion_innovacion = px.bar(
    respuestas_raw,
    x='sector',
    y='inversion_innovacion',
    color='sector',  # Use 'entidad' as the color variable
    labels={'sector': 'Sector', 'inversion': 'Inversion Innovación', 'entidad': 'Entidad'},
    title="Presupuesto de inversión para innovación por Sector",
    hover_data=['entidad']  # Show 'entidad' on hover
)

general = px.bar(
    respuestas_raw,
    x='sector',
    y='general',
    color='sector',  # Use 'entidad' as the color variable
    labels={'sector': 'Sector', 'general': 'General', 'entidad': 'Entidad'},
    title="Presupuesto general por Sector",
    hover_data=['entidad']  # Show 'entidad' on hover
)

general_innovacion = px.bar(
    respuestas_raw,
    x='sector',
    y='general_innovacion',
    color='sector',  # Use 'entidad' as the color variable
    labels={'sector': 'Sector', 'general_innovacion': 'General Innovación', 'entidad': 'Entidad'},
    title="Presupuesto general para innovación por Sector",
    hover_data=['entidad']  # Show 'entidad' on hover
)
```

```python
funcionamiento.update_layout(
    barmode='stack',  # Stack the bars
    width=800,  # Set the width to 800
    height=500,  # Set the height to 500
    xaxis={'title': '', 'showticklabels': False},  # Hide x-axis title and tick labels
)
funcionamiento.show()
funcionamiento.write_html(f'{ruta}/graphs/funcionamiento.html')

funcionamiento_innovacion.update_layout(
    barmode='stack',  # Stack the bars
    width=800,  # Set the width to 800
    height=500,  # Set the height to 500
    xaxis={'title': '', 'showticklabels': False},  # Hide x-axis title and tick labels
)
funcionamiento_innovacion.show()
funcionamiento_innovacion.write_html(f'{ruta}/graphs/funcionamiento_innovacion.html')

inversion.update_layout(
    barmode='stack',  # Stack the bars
    width=800,  # Set the width to 800
    height=500,  # Set the height to 500
    xaxis={'title': '', 'showticklabels': False},  # Hide x-axis title and tick labels
)
inversion.show()
inversion.write_html(f'{ruta}/graphs/inversion.html')

inversion_innovacion.update_layout(
    barmode='stack',  # Stack the bars
    width=800,  # Set the width to 800
    height=500,  # Set the height to 500
    xaxis={'title': '', 'showticklabels': False},  # Hide x-axis title and tick labels
)
inversion_innovacion.show()
inversion_innovacion.write_html(f'{ruta}/graphs/inversion_innovacion.html')

general.update_layout(
    barmode='stack',  # Stack the bars
    width=800,  # Set the width to 800
    height=500,  # Set the height to 500
    xaxis={'title': '', 'showticklabels': False},  # Hide x-axis title and tick labels
)
general.show()
general.write_html(f'{ruta}/graphs/general.html')

general_innovacion.update_layout(
    barmode='stack',  # Stack the bars
    width=800,  # Set the width to 800
    height=500,  # Set the height to 500
    xaxis={'title': '', 'showticklabels': False},  # Hide x-axis title and tick labels
)
general_innovacion.show()
general_innovacion.write_html(f'{ruta}/graphs/general_innovacion.html')
```

```python
import plotly.graph_objects as go

sector_totals_general = respuestas_raw.groupby('sector')[['general']].sum().reset_index().sort_values(by='sector')
sector_totals_general_innovacion = respuestas_raw.groupby('sector')[['general_innovacion']].sum().reset_index().sort_values(by='sector')

fig_general = go.Figure(data=[
    go.Bar(
        x=sector_totals_general['sector'],
        y=sector_totals_general['general'],
        name='Presupuesto General',
        marker_color='blue',
    ),
    go.Bar(
        x=sector_totals_general_innovacion['sector'],
        y=sector_totals_general_innovacion['general_innovacion'],
        name='Presupuesto General Innovación',
        marker_color='orange',
    )
])

fig_general.update_layout(
    barmode='overlay',
    title="Presupuesto General vs. Presupuesto General Innovación",
    width=800,
    height=500,
    xaxis={'title': '', 'showticklabels': False},
    yaxis={'title': '', 'showticklabels': True},
)

fig_general.write_html(f'{ruta}/graphs/general_vs_general_inno.html')
fig_general.show()
```
